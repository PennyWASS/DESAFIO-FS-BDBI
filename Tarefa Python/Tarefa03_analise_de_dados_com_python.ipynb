###Criando os dados a partir das informações obtidas anteriormente

```python
#Agrupando a partir da quantidade de funcionários em cada setor
df2 = df.groupby('Department')[['EmployeeCount']].count()
#Apresentando a nova tabela
df2
```

```python
#Criação do gráfico
df2.plot(kind='bar')
#Estilizando o gráfico:
plt.title('Quantidade de funcionários em cada setor')
plt.xlabel('Departamento')
plt.ylabel('Quantidade de funcionários')
```

```python
#Agrupando a partir da quantidade de funcionários para cada nível de escolaridade:
df3 = df.groupby('Education')[['EmployeeCount']].count()
#Apresentando a nova tabela:
df3
```

```python
#Criação do gráfico:
df3.plot(kind='line', color = '#6e3c1d')
#Estilizando o gráfico:
plt.title('Quantidade de funcionários para cada nível de escolaridade')
plt.xlabel('Nível de escolaridade')
plt.ylabel('Quantidade de funcionários')
```

```python
#Agrupando a partir da quantidade de funcionários que tem ou não algum atrito na empresa:
df4 = df.groupby('Attrition')[['EmployeeCount']].count()
#Apresentando a nova tabela:
df4
```

```python
#Criação do gráfico:
df4.plot(kind='pie', subplots=True, colors=['#81f542','#f4ff24'])
#Estilizando o gráfico:
plt.title('Quantidade de funcionários que tem ou não algum atrito na empresa')
plt.ylabel('Funcionários com algum atrito?')
```
